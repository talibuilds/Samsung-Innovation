```python
import numpy as np
a = np.array([[3,4,1], [0, 5, 2]])
b = np.array([[2,0,0], [1, 0, 5], [4, 5, 0]])

c = np.dot(a, b)

print("Matrix A:\n", a)
print("Matrix B:\n", b)
print("Result AXB:\n", c)
```

```python
import numpy as np
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

c = a+b
d = a- (1/2)*b
e = 5*a

print("Matrix A:\n", a)
print("Matrix B:\n", b)
print("Addition A+B:\n", c)
print("A-(1/2)XB:\n", d)
print("5XA:\n", e)
```

```python
import numpy as np

a = np.array([[2, 3, 1], [4, 5, 1]])
b = np.array([[-1, 2], [4, -2], [3, 6]])
print("AB:\n", np.dot(a, b))
print("BA:\n", np.dot(b, a))
```

```python
import numpy as np
x = np.array([[1,2,6], [1,4,5],[1,6,7]])
y = np.array([[50], [60], [80]])

beta = np.linalg.inv(x.T @ x) @ x.T @ y
print("Design Matrix X:\n", x)
print("Design Matrix Y:\n", y)
print("Coefficient (Beta)\n:", beta)
```

```python
import numpy as np

a = np.array([[0, 0, 3], [0, 0, 0], [1, 0, 0]])

zero_count = np.sum(a==0)
total_element = a.size

print("Zero Elements:\n", zero_count)
print("Total Elements:\n", total_element)

if zero_count > (total_element/2) :
    print("\nSparse Matrix")

else:
    print("Not a Sparse Matrix")
```

```python
import numpy as np

matrix = np.array([[0, 0, 3], [0, 0, 0], [1, 0, 0]])

rows, cols = matrix.shape
sparse_rep = []

for i in range(rows):
    for j in range(cols):
        sparse_rep.append((i, j, matrix[i][j]))

print("Sparse Representaion Row Column Values:\n")
for item in sparse_rep:
    print(item)
```
