```python
# AP
a = 2
d = 3
n = 50
seq = []
for i in range(n):
    term = a + i*d
    seq.append(term)
print(seq)

# Other way
sequence= list(range(3, 30, 3))  # AP with Cd 3
print(sequence)
```

```python
# Finding the nth term
def find_n_term(a, d, n):
    return a + (n-1)*d

print(find_n_term(2, 5, 7))
```

```python
# Sum of AP

def sum_of_ap(a, d, n):
    return (n/2)*(2*a + (n-1)*d)

print(sum_of_ap(2, 3, 10))

# OTher way
a, d, n = 2, 3, 10
total = sum(a + i*d for i in range(n))
print(total)
```

```python
# Check if the List in AP

def is_ap(nums):
    if(len(nums)) <2:
        return True

    diff = nums[1] - nums[0]

    for i in range(1, len(nums) -1):
        if nums[i+1] -nums[i] !=diff:
            return False
    return True

print(is_ap([5, 7, 15]))
print(is_ap([5, 7, 9]))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create time series
time_index = pd.date_range(start="2026-03-25 00:00:00", periods=10, freq="d")
temperature = np.random.randint(20, 35, size=10)

df = pd.DataFrame({
    'Timestamp': time_index,
    'temperature': temperature
})

df.set_index('Timestamp', inplace=True)

print("Time Series Data:\n")
print(df)

# Basic stats
print("\nMean:", df["temperature"].mean())
print("Max:", df["temperature"].max())
print("Min:", df["temperature"].min())

# Trend (difference)
df["Trends"] = df["temperature"].diff()

print("\nWith Trends:\n")
print(df)

# 📊 Plot temperature and trend
plt.figure(figsize=(10,5))

plt.plot(df.index, df["temperature"], marker='o', label="Temperature")
plt.plot(df.index, df["Trends"], marker='x', label="Trend (diff)")

plt.title("Temperature and Trend Over Time")
plt.xlabel("Time")
plt.ylabel("Values")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import numpy as np
import pandas as pd

hrs = np.array([1, 2, 3, 5, 6, 8, 9, 10])
scaled = (hrs -hrs.min()) / (hrs.max() -hrs.min())

bins = np.linspace(hrs.min(), hrs.max(), 4)

labels = ["Low", "Medium", "High"]

categories = pd.cut(hrs, bins =bins, labels =labels, include_lowest = True)

df = pd.DataFrame({
    "Hours" :hrs,
    "Scaled" : scaled,
    "Category" : categories
})
print(df)
```
