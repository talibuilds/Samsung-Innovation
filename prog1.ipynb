```python
import numpy as np
import matplotlib.pyplot as plt

theta = np.linspace(0, 8*np.pi, 1000)
r = theta

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.scatter(x, y, c=theta, cmap='hsv', s=5)
plt.axis('off')
plt.show()
```
