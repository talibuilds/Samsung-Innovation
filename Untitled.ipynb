```python
import numpy as np

salaries = np.array([25, 30, 28, 27, 29, 31, 5])

mean_sal = np.mean(salaries)

deviation = np.abs(salaries -mean_sal)
threshold = 3*np.mean(deviation)
outliers = salaries[deviation > threshold]

print(salaries)
print(mean_sal)
print(deviation)
print(threshold)
print(outliers)
```

```python
# KNN - k nearest Neighbour
import numpy as np
from numpy.linalg import norm

grp_a = np.array([[40, 50], [42, 48], [38, 52]])
grp_b = np.array([[70, 80], [72, 78], [68, 82]])

new_student = np.array([45, 55])
dist_a = [norm(new_student -point) for point in grp_a]
dist_b = [norm(new_student -point) for point in grp_b]


print("Distance from group A: ", dist_a)
print("Distance from group B: ", dist_b)

if min(dist_a) < min(dist_b):
    print("Belongs to Group A")
else:
    print("Belongs to Group B")
```
