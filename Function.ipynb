```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1001)
y = 3*x +1
plt.plot(x, y, label='f(x) = 3x +1')
plt.xlabel("Domain")
plt.ylabel("Range")
plt.title("Function demo")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1001)
g = x**2
f_of_g = 3*(g) +1


plt.plot(x, g, label='g(x) = x^2')
plt.plot(x, f_of_g, label='f(g(x)) = 3x^2 +1')
plt.xlabel("x")
plt.ylabel("y")
plt.title("composite func")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1001)
y = 3*x +1

y_inv = (x-1)/3




plt.plot(x, y, label='f(x) = 3x +1')
plt.plot(x, y_inv, label='f^-1(x) = (x-1) /3')
plt.plot(x, x, linestyle = "--", label = "y=x")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Function and inverse function")
plt.grid()
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
url = "https://raw.githubusercontent.com/selva86/datasets/master/Advertising.csv"
data = pd.read_csv(url)
x = data["TV"]
y = data["sales"]
coeff = np.polyfit(x, y, 2)
poly = np.poly1d(coeff)

x_new = np.linspace(x.min(), x.max(), 100)
y_new = poly(x_new)
plt.scatter(x, y)
plt.plot(x_new, y_new, color="red")
plt.title("Regression model")
plt.xlabel("TV advertising budget")
plt.ylabel("Sales")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.linspace(-2, 5, 100)
x2 = np.linspace(-5, 5, 100)
y1 = 2 **x1
y2 = 0.5 **x2

plt.plot(x1, y1)
plt.plot(x2, y2)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Exponential Growth and Decay")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.1, 10, 100)
y = np.log(x)

a = np.linspace(0.1, 5, 100)
b = np.exp(x)

plt.plot(x, y)
plt.plot(a, b)
plt.title("Logarithmic Function: f(x) = log(x)")
plt.xlabel("x")
plt.ylabel("log(x)")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

days = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)

cases = np.array([5, 9, 20, 40, 85, 170])

log_cases = np.log(cases)
model = LinearRegression()
model.fit(days, log_cases)

predicted_log = model.predict(days)
predicted_cases = np.exp(predicted_log)

plt.scatter(days, cases, label="Actual Data")
plt.plot(days, predicted_cases, label="Exponential Fit")
plt.title("Exponential Growth")
plt.xlabel("Days")
plt.ylabel("Cases")
plt.legend()
plt.show()
```
