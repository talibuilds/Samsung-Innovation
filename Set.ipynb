```python
AI = {"S1", "S2", "S3"}
DS= {"S4", "S2", "S3"}
ML = {"S3", "S5", "S6"}

at_least_two = (AI &ML) | (AI &  DS) | (DS &ML)
only_ai = AI - (ML | DS)

print(at_least_two)
print(only_ai)
```

```python
predicted = {"mail1", "mail2", "mail3"}
actual = {"mail2", "mail3", "mail5"}

print("True Positive: ", predicted & actual)
print("False Positive", predicted -actual)
print("false Negative", actual -predicted)
```

```python
visited = {"A", "B", "C", "D"}
purchased = {"B", "C", "E", "F"}

print("Visited and Purchased: ", visited & purchased)
print("Unique users: ", visited | purchased)
print("Who did not purchased: ", visited - purchased)
print("Purchased but did not visited: ", purchased -visited)
```

```python
import pandas as pd

visited_data = {
    "Cust_ID" : ["C1", "C2", "C3", "C4", "C5", "C6"],
    "visit_count" : [5, 3, 2, 8, 7, 4]
}
Purchased_data = {
    "Cust_ID" : ["C3", "C4", "C5", "C7"],
    "Purchase_Amount" : [200, 150, 300, 250]
}

visited_df = pd.DataFrame(visited_data)
purchased_df = pd.DataFrame(Purchased_data)
print("Visited: \n", visited_df)
print("\nPurchased \n", purchased_df)


visited_user = set(visited_df["Cust_ID"])
purchased_user = set(purchased_df["Cust_ID"])
print("Visited users", visited_user)
print("Purchased users: ", purchased_user)



common_users = visited_user & purchased_user
print("Common Users: ", common_users)
print("Users who did not purchased", visited_user - purchased_user)
print("Purchased but not visited", purchased_user - visited_user)
print("All unique users: ", visited_user | purchased_user)

p = visited_user
a = purchased_user

print("\nClassification analysis")
print("True Positive ", p & a )
print("False Positive", p-a)
print("False Negative", a-p)


common_df = pd.merge(visited_df, purchased_df, on="Cust_ID", how="inner")
left_df = pd.merge(visited_df, purchased_df, on="Cust_ID", how="left")
non_buyers = left_df[left_df["Purchase_Amount"].isna()]

print("Common Users: ", common_df)
print("\nVisiters who did not purchased: ", non_buyers)
```
